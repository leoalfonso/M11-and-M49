# ðŸ Period 7: Data Visualization

We have loaded, cleaned, and calculated statistics for our precipitation data. Now, we will visualize it.

A plot or graph can often reveal insights far more quickly than a table of numbers. For example:
* Is there a seasonal trend?
* Are the rainfall events clustered together?
* How skewed is the distribution of rainfall? (i.e., many small events vs. a few large ones)

We will use two primary plotting libraries:
1.  **Matplotlib:** The foundational plotting library in Python. We use its `pyplot` module, typically imported as `plt`.
2.  **Seaborn:** A higher-level library built on Matplotlib that simplifies creating common, statistically-attractive plots. We typically import it as `sns`.
