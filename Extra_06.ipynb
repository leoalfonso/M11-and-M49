**Challenge: Antecedent Precipitation Index (API)**

The Antecedent Precipitation Index (API) is a common metric in hydrology to estimate soil moisture. It's a weighted sum of previous rainfall. A simple version is:
$$API_i = k \times API_{i-1} + P_i$$
Where:
* $API_{i-1}$ = The API from the *previous* day
* $k$ = A recession constant (e.g., $k=0.9$), representing 90% of moisture being retained.
* $P_i$ = Precipitation on day $i$

This is difficult to do in a simple vectorized way, so we will use `.ewm()` (Exponentially Weighted Moving average) which is built for this.

**Task:**
1.  Reload the *original* `precipitation_data.csv` and set the `Date` index.
2.  Use the `.ewm()` (Exponentially Weighted Moving) method to calculate the API.
3.  Use the following syntax: `df['API'] = df['Precipitation_mm'].ewm(alpha=0.1, adjust=False).mean()`
    * `alpha=0.1` is the smoothing factor ($1-k$).
4.  Print the `Precipitation_mm` column and the new `API` column side-by-side to compare them.
