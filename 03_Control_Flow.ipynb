# ðŸ Period 3: Control Flow (Conditionals & Loops)

In computational modeling, we rarely execute a single linear set of instructions. We require **control flow** to handle decisions and repetitions.

* **Conditionals (`if`):** Essential for checking regulatory compliance, classifying data (e.g., flood stages), or directing the model based on input parameters.
* **Loops (`for`):** Critical for automating repetitive, large-scale calculations, such as processing daily measurements over decades or iterating through every element of a network model.

## 1. Conditional Logic (`if`, `elif`, `else`)

Conditionals rely on **Boolean logic** (True/False) evaluated by **logical operators** (e.g., `>`, `<`, `==`, `!=`).

**Key Syntax Feature:** Python uses **indentation** (usually four spaces) to define the block of code that belongs to an `if` or `elif` statement. This structure is non-negotiable.
