**Challenge: Classifying Flow Regime**

The **Reynolds Number ($Re$)** is a dimensionless quantity used to predict flow patterns. For pipe flow, it is defined as:
$$Re = \frac{V D}{\nu}$$
Where:
* $V$ = Mean velocity ($\text{m/s}$)
* $D$ = Pipe diameter ($\text{m}$)
* $\nu$ (nu) = Kinematic viscosity of the fluid ($\text{m}^2/\text{s}$). For water at 20Â°C, $\nu \approx 1.004 \times 10^{-6} \text{ m}^2/\text{s}$.

**Task:**
1.  You are given a list of mean velocities (in $\text{m/s}$): `velocities = [0.01, 0.1, 0.5, 1.5, 3.0]`.
2.  Set `D = 0.5` m and `nu = 1.004e-6`.
3.  Write a `for` loop to iterate through each velocity.
4.  Inside the loop, calculate the Reynolds Number for that velocity.
5.  Use an `if/elif/else` block to classify the flow based on $Re$:
    * $Re < 2300$: "Laminar Flow"
    * $2300 \le Re < 4000$: "Transitional Flow"
    * $Re \ge 4000$: "Turbulent Flow"
6.  Print the velocity, the calculated $Re$, and the flow classification for each case.
