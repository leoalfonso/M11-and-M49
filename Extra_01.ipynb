**Challenge: Manning's Equation for Velocity**

The average velocity $V$ (in $\text{m/s}$) in an open channel can be estimated using Manning's Equation:
$$V = \frac{1}{n} R_h^{2/3} S^{1/2}$$
Where:
* $n$ = Manning's roughness coefficient (e.g., `0.015` for finished concrete)
* $R_h$ = Hydraulic Radius (in meters)
* $S$ = Channel slope (dimensionless, e.g., `0.001` for $1\text{m}$ drop per $1000\text{m}$)

**Task:**
Write a Python script that:
1.  Defines variables for $n$, $R_h$, and $S$.
2.  Calculates the velocity $V$.
3.  Given this velocity $V$ and a cross-sectional area $A = 2.5 \text{ m}^2$, calculate the discharge $Q$ (in $\text{m}^3/\text{s}$), using the continuity equation $Q = V \times A$.
4.  Print the final discharge $Q$.
