## 1. Review of Core Data Types

Last period, we worked with **numeric** data (integers and floats) and basic arithmetic. A solid understanding of Python's fundamental data types is crucial for defining variables correctly in engineering models.

| Data Type | Description | Hydrology/Hydraulics Example |
| :---: | :--- | :--- |
| **int** | Whole numbers (integers). | The number of pipe segments (e.g., 12). |
| **float** | Numbers with a decimal point (floating-point precision). | A flow rate $Q$ (e.g., 2.345 $\text{m}^3/\text{s}$). |
| **str** | Text data (strings), enclosed in quotes. | A water body name (e.g., "Nile River"). |

```python
# Task 1.1: Define variables using the three core data types.

# 1. Integer: Number of monitoring wells in a field site
num_wells = 7

# 2. Float: Groundwater depth measurement (in meters)
gw_depth_m = 15.68

# 3. String: Site Identifier
site_id = "GW-Piezometer_A"

# Use an f-string to combine and print these variables in a descriptive sentence.
print(f"Monitoring Site: {site_id} has {num_wells} wells, with the latest measurement at {gw_depth_m} meters depth.")
```
